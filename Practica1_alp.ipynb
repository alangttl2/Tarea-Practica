```python
########################## Alan Ledezma Puente #################################

def pila(entrada):
    """
    Función que procesa una expresión y genera la notación polaca inversa.
    """
    lst_1 = len(entrada)
    file.writelines("Inicia valores dentro de la función def polaca \n")

    # Listas que servirán para guardar signos u operadores.
    lst_operadores = []  # Lista donde se guardarán los números
    lst_signo = []  # Lista donde se guardarán los signos
    lst_parentesis = []  # Lista que junta signos y operadores
    lst_sol = []  # Lista pivote o auxiliar de nuestras otras listas
    lst_polaca = []  # Pila que será la salida de la entrada aritmética

    # Seteamos las variables a 0 para después usarlas con el contador
    contador_1 = 0  # Este sera el apuntador
    parentesis_izq = 0
    # Variables de signos:
    var_minus = 0
    var_sum = 0
    var_div = 0
    var_mult = 0
    var_exp = 0
    #Son de valor 0 para indicar que iniciara así la variable
    conparentesis = lst_1 - 1

    # Comienza a generar la pila ingresada
    file.writelines("Lectura de la pila ingresada \n")

    # Barrido de toda la pila para acomodar signos y operadores.
    while contador_1 <= lst_1 - 1:

      # Inico de casos dentro del barrido de la pila
        if entrada[contador_1] == "(":
            # Caso para operadores y signos antes del paréntesis.
            if var_exp == 1:
                lst_signo.append("**")
            if var_mult == 1:
                lst_signo.append("*")
            if var_div == 1:
                lst_signo.append("/")
            if var_sum == 1:
                lst_signo.append("+")
            if var_minus == 1:
                lst_signo.append("-")
            # Estara activo cuando se encuentra x signo y lo agregara a
            # lst_signo.

            # Concatenamos listas en la lista auxiliar osea en lst_sol
            lst_sol = lst_operadores + lst_signo
            file.writelines("Ordena cuando hay y no hay un paréntesis \n")

            # Vacíamos las listas para que no se queden valores pues es un ciclo
            lst_polaca.append(lst_sol)
            lst_signo.clear()
            lst_operadores.clear()
            # Fin de caso antes de parentesis

            # Para casos entre de parentesis
            parentesis_izq = 1
            if contador_1 == 0:
                conparentesis = 1
            # Seteamos las variables a 0 para evitar problemas de valores
            var_exp = 0
            var_mult = 0
            var_div = 0
            var_sum = 0
            var_minus = 0
            contador_1 += 1 # Apuntador de nuevo porque va a leer el siguiente
            # elemento

        #Evalua cuando tu parentesis izq "(" esta activo
        elif entrada[contador_1] == ")":
            parentesis_izq = 0  # Fin de paréntesis pues "(" = 0
            if var_exp == 1:
                lst_signo.append("**")
            if var_mult == 1:
                lst_signo.append("*")
            if var_div == 1:
                lst_signo.append("/")
            if var_sum == 1:
                lst_signo.append("+")
            if var_minus == 1:
                lst_signo.append("-")

            # Vacíamos las listas para que no se queden valores pues es un ciclo
            lst_parentesis = lst_operadores + lst_signo
            lst_signo.clear()
            lst_operadores.clear()
            #Fin caso entre parentesis

           #Caso para despúes del parentesis.
            var_exp = 0
            var_mult = 0
            var_div = 0
            var_sum = 0
            var_minus = 0
            contador_1 += 1 # Apuntador que lee que leera el sig. elemento

        elif parentesis_izq == 1:
            # Dentro de los paréntesis
            num = entrada[contador_1]
            if num.isnumeric():
                lst_operadores.append(entrada[contador_1])
            else:
                # Es lo mismo que los otros casos en cuanto el "switch" y
                # jerarquización de signos pero aplicado a los números fuera del
                # parentesis
                if entrada[contador_1] == "-":
                    var_minus = 1
                if entrada[contador_1] == "+":
                    var_sum = 1
                if entrada[contador_1] == "/":
                    var_div = 1
                if entrada[contador_1] == "*" and entrada[contador_1 + 1] != \
                    "*" and entrada[contador_1 - 1] != "*":
                    var_mult = 1
                if entrada[contador_1] == "*" and entrada[contador_1 - 1] == \
                    "*":
                    var_exp = 1
                    #Valida que el elemento anterior no sea un "*"

            contador_1 += 1
            # Evalua si el parentesis esta activo o esta apagado
            # Fin caso para después parentesis

        elif parentesis_izq == 0:
            # Caso para operadores y signos fuera de los paréntesis
            num = entrada[contador_1]
            if num.isnumeric():
                lst_operadores.append(entrada[contador_1])
                # Si es un número lo acomodara para con el apuntador(contador_1)
                # en tu lista de operadores
            else:
                if entrada[contador_1] == "-":
                    var_minus = 1 # Es = 1 para denotar que ya hay un signo así
                if entrada[contador_1] == "+":
                    var_sum = 1
                if entrada[contador_1] == "/":
                    var_div = 1
                if entrada[contador_1] == "*" and entrada[contador_1 + 1] != "*" \
                        and entrada[contador_1 - 1] != "*":
                    var_mult = 1
                if entrada[contador_1] == "*" and entrada[contador_1 - 1] == "*":
                    var_exp = 1

            lst_sol = lst_operadores + lst_signo
            contador_1 += 1
            # Fin del if
#Fin de While

    if var_exp == 1:
        lst_signo.append("**")
    if var_mult == 1:
        lst_signo.append("*")
    if var_div == 1:
        lst_signo.append("/")
    if var_sum == 1:
        lst_signo.append("+")
    if var_minus == 1:
        lst_signo.append("-")
    # Esta iterado aquí pq es para identificar los ultimos signos donde se
    # escendio el "switch" de los signos.

    if entrada[conparentesis] == ")":
        lst_polaca += lst_parentesis
    else: # En otro caso que no se encuentre ")"
        lst_sol.append(lst_signo)
        lst_polaca += lst_parentesis + lst_sol

    # Para eliminar caracteres no deseados
    formatted_output = ''.join(
        str(lst_polaca).replace('[', '').replace(']', '').replace("'", '').\
        replace(",", '')
    )
    return formatted_output  # Lo que te debe de regresar como pila final.


# Main
if __name__ == '__main__':
    entrada = input("Introduzca su operación: ")
    file = open("AlanLedezmaPuente_Practica1.txt", "w") # Creación del archivo
    file.writelines("Inicia programa leyendo una cadena de entrada \n")

    result_polaca = pila(entrada)
    print(result_polaca) # Salida en forma polaca inversa de tu expresion
    # ingresada aritmetica

    file.close() # Fin del archivo creado
```
